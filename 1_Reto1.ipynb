```python
import numpy as np #Posee funciones matemáticas
import pandas as pd #Comunicación con el excel

```

1. Se carga la base de datos

```python
nxl='/content/1. BD1_Longitudinal Survey_Int.xlsx'
XDB=pd.read_excel(nxl)
XDB.head(100)
```

Seleccionar las variables

```python
XDB=XDB[['Age','Urban','Mother_Edu','Father_Edu','FamilySize','Self_Esteem','Height','Weight','Income']]
XDB.head(10)
```

```python
# Eliminar filas con al menos un valor faltante
XDB = XDB.dropna()
XDB.head(10)


```

```python
total_respondents = len(XDB)
print(f"Número total de personas que respondieron: {total_respondents}")
```

```python
# Porcentaje de cada categoría en "Urban"
categories = XDB['Urban'].value_counts(normalize=True) * 100
print('Porcentaje de datos en cada categoría:\n', categories)
#1.0 → 79.16%: Significa que el 79.16% de las personas en la base de datos viven en áreas urbanas.
#0.0 → 20.83%: Indica que el 20.83% de las personas viven en áreas rurales.
```

Se separan las variables de entrada y salida

```python
#Separan las variables de entrada y salida
XD=XDB[['Age','Mother_Edu','Father_Edu','FamilySize','Self_Esteem','Height','Weight','Income']] #Variables de entrada
YD=XDB[['Urban']] #Variable de salida-Variable Pronóstico
```

Se comienza la implementación del modelo

```python
# Intervalo de la Edad
age_min = XDB['Age'].min()
age_max = XDB['Age'].max()
age_interval = (age_min, age_max)

# Intervalo de la Altura
height_min = XDB['Height'].min()
height_max = XDB['Height'].max()
height_interval = (height_min, height_max)

# Intervalo del Peso
weight_min = XDB['Weight'].min()
weight_max = XDB['Weight'].max()
weight_interval = (weight_min, weight_max)

# Imprimir los resultados
print(f"Intervalo de la Edad: {age_interval}")
print(f"Intervalo de la Altura: {height_interval}")
print(f"Intervalo del Peso: {weight_interval}")


# Filtrar los datos para el área urbana (Urban = 1)
datos_urbanos = XDB[XDB['Urban'] == 1]

# Calcular la media y la desviación estándar de la altura en el área urbana
media_altura_urbana = datos_urbanos['Height'].mean()
desviacion_estandar_altura_urbana = datos_urbanos['Height'].std()

# Calcular los límites superior e inferior
limite_superior = media_altura_urbana + desviacion_estandar_altura_urbana
limite_inferior = media_altura_urbana - desviacion_estandar_altura_urbana

# Imprimir los resultados
print("Límite superior de la altura en el área urbana:", limite_superior)
print("Límite inferior de la altura en el área urbana:", limite_inferior)

# Filtrar los datos para el área rural (Urban = 0)
datos_rurales = XDB[XDB['Urban'] == 0]

# Calcular la media y la desviación estándar del peso en el área rural
media_peso_rural = datos_rurales['Weight'].mean()
desviacion_estandar_peso_rural = datos_rurales['Weight'].std()

# Calcular los límites superior e inferior
limite_superior = media_peso_rural + desviacion_estandar_peso_rural
limite_inferior = media_peso_rural - desviacion_estandar_peso_rural

# Imprimir los resultados
print("Límite superior del peso en el área rural:", limite_superior)
print("Límite inferior del peso en el área rural:", limite_inferior)
```

```python
from sklearn.naive_bayes import GaussianNB

mnb=GaussianNB() #Aloja el modelo
mnb.fit(XD,YD) #Entrena el modelo

#Obtenemos la información del modelo
u=mnb.theta_ #Obtenemos la media de las variables
print("La media de las variables es:\n",u)
var=mnb.var_ #Se calculan las varianzas
sigma=np.sqrt(var) #Se calcula la desviación estándar
print("La desviación estándar de los ingresos es:\n",sigma)

#Se determinan los límites superiores e inferiores de los datos
LS=u+sigma
LI=u-sigma
print("El límite superior de las variables es:\n",LS)
print("El límite inferior de las variables es:\n",LI)

#Evaluamos un solicitante
XDS=[[34,12,15,2,24,69,160,48000]] #Como un valor de entrada-Ingresos
ydp=mnb.predict(XDS) #Se obtiene la prediccion
print("La predicción del encuestado es:\n", ydp)

if ydp==0:
  print("El solicitante vive en zona rural")
if ydp==1:
  print("El solicitante vive en zona urbana")
```

Evaluamos el desempeño del modelo

```python
from re import S
from sklearn.metrics import confusion_matrix

#Haga un barrido de los datos
ydp2=mnb.predict(XD) #Haga el pronostico para los datos base
cm=confusion_matrix(YD,ydp2) #Realiza la matriz de confusión
print("La matriz de confusión es:\n",cm)
VN=cm[0,0]
FP=cm[0,1]
FN=cm[1,0]
VP=cm[1,1]

#Métricas desempeño
#1.Exactitud
Ex=(VP+VN)/(VP+VN+FP+FN) #Exactitud: comportamiento general urbano
print("La exactitud del modelo es:\n",Ex)

TE=(FN+FP)/(VP+VN+FP+FN) #Tasa de error: comportamiento general rural
print("La tasa de error del modelo es:\n",TE)

#2 Sensibilidad

Se=VP/(VP+FN) #Sensibilidad: comportamiento correcto urbano
print("La sensibilidad del modelo es:\n",Se)

#3 Especifidad

Sp=VN/(VN+FP) #Especificidad: comportamiento correcto rural
print("La especificidad del modelo es:\n",Sp)

#4 La precisión
Pr=VP/(VP+FP) #Precisión: comportamiento correcto urbano #Sobre el total de los precedidos

print("La precisión del modelo es:\n",Pr)

#5 La precisión negativa
PrNeg=VN/(VN+FN) #Precisión negativa: comportamiento correcto rural #Sobre el total de los no precedidos

print("La precisión negativa del modelo es:\n",PrNeg)



```

Con base en la información suministrada, se desarrolló un modelo para determinar si una persona reside en una zona urbana o rural. Para ello, se consideraron variables socioeconómicas como edad, educación de los padres, estatura, peso, ingresos, escala de autoestima y tamaño de la familia.

La base de datos original contenía un total de 12.686 registros. Sin embargo, tras un proceso de depuración para eliminar las respuestas incompletas, se obtuvo un conjunto final de 5.948 datos, de los cuales el 79,16% corresponde a personas que viven en zonas urbanas y el 20,83% en zonas rurales.


El modelo alcanzó una **precisión** promedio del 79,08%, lo que indica un desempeño aceptable en la clasificación de la zona de residencia para la zona urbana. Sin embargo, la **especificidad** fue del 0,24%, lo que refleja una mayor dificultad en la identificación de residencia rurales. A pesar de esto, el **exactitud** del modelo se mantuvo cercano al 79%, evidenciando un criterio sólido en la determinación residencia para la zona urbana. Lo que nos da entender que el modelo es excelente para la clasificación de la zona de residencia para la zona urbana.
